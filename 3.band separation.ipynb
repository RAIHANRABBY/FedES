```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import mne
import pywt
```

```python
folder = 'E:/Data WareHouse/1sec/Epileptic Seizure/equalized'
epochs_path = [os.path.join(folder,i) for i in os.listdir(folder) if i[-3:]=='fif']
epochs_path
```

# theta

```python
theta= [mne.read_epochs(i).pick_types(eeg=True).filter(4,8) for i in epochs_path]
labels = [mne.read_epochs(i).events[:,2] for i in epochs_path]
```

```python
theta[0].events
```

```python
theta_data = [i.get_data() for i in theta]
```

```python
group = [[i]*len(j) for i,j in enumerate(theta_data)]
```

```python
X=np.vstack(theta_data)
Y=np.hstack(labels)
group= np.hstack(group)
print(X.shape,Y.shape,group.shape)
```

```python
np.savez("E:/Data WareHouse/1sec/Epileptic Seizure/equalized/equalized data/theta.npz",X = X,Y = Y,G = group)
```
